```python
import time
import json
import pygetwindow as gw
import pyautogui
from tkinter import Tk, messagebox

# File to store allowed sites/titles
ALLOWED_FILE = "allowed_tabs.json"

def load_allowed():
    try:
        with open(ALLOWED_FILE, "r") as f:
            return json.load(f)
    except FileNotFoundError:
        return []

def save_allowed(allowed):
    with open(ALLOWED_FILE, "w") as f:
        json.dump(allowed, f)

def ask_permission(title):
    root = Tk()
    root.withdraw()  # Hide main window
    answer = messagebox.askyesno("Tab Detected", f"Do you want to keep this tab open?\n\nTitle: {title}")
    root.destroy()
    return answer

def close_tab(window):
    window.activate()
    time.sleep(0.5)
    pyautogui.hotkey('ctrl', 'w')  # Close tab

def main_loop():
    print("Monitoring browser tabs...")
    known_tabs = set()
    allowed_titles = load_allowed()

    while True:
        all_windows = gw.getAllTitles()
        for title in all_windows:
            if title and title not in known_tabs and any(browser in title.lower() for browser in ['chrome', 'firefox', 'edge']):
                known_tabs.add(title)
                print(f"New tab detected: {title}")

                if any(allowed.lower() in title.lower() for allowed in allowed_titles):
                    print("Previously allowed, skipping...")
                    continue

                if ask_permission(title):
                    allowed_titles.append(title)
                    save_allowed(allowed_titles)
                    print(f"Tab allowed: {title}")
                else:
                    win = gw.getWindowsWithTitle(title)
                    if win:
                        close_tab(win[0])
                        print(f"Tab closed: {title}")
        time.sleep(3)

if __name__ == "__main__":
    main_loop()
```
