* ### Can we accurately classify patient sentiment (positive/negative/neutral) from drug reviews using NLP techniques?


```python

# Create positive and negative word dictionaries
positive_words = {
    'amazing': 2, 'excellent': 2, 'lifesaver': 2, 'effective': 1.5,
    'great': 1.5, 'perfect': 2, 'love': 1.5, 'best': 2, 'wonderful': 2,
    'good': 1, 'works': 1, 'helpful': 1, 'well': 1, 'helped': 1,
    'relief': 1.5, 'recommend': 1.5, 'happy': 1.5, 'improved': 1.5
}

negative_words = {
    'terrible': -2, 'awful': -2, 'horrible': -2, 'useless': -2,
    'bad': -1, 'poor': -1.5, 'waste': -1.5, 'worse': -1.5,
    'side effects': -2, 'nausea': -1, 'pain': -1, 'dizzy': -1,
    'stomach': -1, 'headache': -1, 'allergic': -2, 'severe': -1.5,
    'stopped': -1, 'quit': -1, 'discontinue': -1.5
}

def simple_sentiment_score(text):
    if pd.isna(text):
        return 0
    
    text_lower = str(text).lower()
    positive_score = 0
    negative_score = 0
    
    for word, weight in positive_words.items():
        if word in text_lower:
            positive_score += weight
    
    for word, weight in negative_words.items():
        if word in text_lower:
            negative_score += weight
    
    total_words = positive_score + abs(negative_score)
    if total_words == 0:
        return 0
    
    final_score = (positive_score - abs(negative_score)) / total_words
    return final_score

def classify_sentiment(score):
    if score > 0.2:
        return 'POSITIVE'
    elif score < -0.2:
        return 'NEGATIVE'
    else:
        return 'NEUTRAL'


# Apply sentiment analysis
drugs['sentiment_score'] = drugs['review'].apply(simple_sentiment_score)
drugs['sentiment_label'] = drugs['sentiment_score'].apply(classify_sentiment)

print("Sentiment Analysis")
drugs[['review', 'rating', 'sentiment_score', 'sentiment_label']].head(10)
```

* ### How accurately can we detect specific emotions beyond positive/negative?

```python
# Create ground truth from ratings (THIS IS MISSING)
drugs['sentiment_true'] = drugs['rating'].apply(
    lambda x: 'POSITIVE' if x >= 8 else ('NEGATIVE' if x <= 4 else 'NEUTRAL')
)

# Now run your complaints code
# Filter negative reviews
negative_reviews = drugs[drugs['sentiment_true'] == 'NEGATIVE']['review'].tolist()
positive_reviews = drugs[drugs['sentiment_true'] == 'POSITIVE']['review'].tolist()

# Complaint keywords
complaint_keywords = {
    'side effects': 0,
    'nausea': 0,
    'stomach': 0,
    'dizzy': 0,
    'headache': 0,
    'weight gain': 0,
    'fatigue': 0,
    'insomnia': 0,
    'diarrhea': 0,
    'vomiting': 0,
    'rash': 0,
    'allergic': 0,
    'does not work': 0,
    'ineffective': 0,
    'terrible': 0,
    'worse': 0,
    'severe': 0
}

# Count complaints
for review in negative_reviews:
    if pd.notna(review):
        review_lower = str(review).lower()
        for keyword in complaint_keywords:
            if keyword in review_lower:
                complaint_keywords[keyword] += 1

# Sort and display
sorted_complaints = sorted(complaint_keywords.items(), key=lambda x: x[1], reverse=True)

print(f"Complaints (from {len(negative_reviews)} negative reviews):")
print(f"{'Complaint Type':<25} {'Count':<10} {'Percentage'}")
print("-" * 60)

complaint_data = []
total_neg = len(negative_reviews)
for complaint, count in sorted_complaints[:10]:
    if count > 0:
        percentage = (count / total_neg) * 100
        print(f"{complaint:<25} {count:<10} {percentage:>6.2f}%")
        complaint_data.append({'Complaint': complaint, 'Count': count, 'Percentage': percentage})

complaint_df = pd.DataFrame(complaint_data)
complaint_df.to_csv('top_complaints.csv', index=False)

# Benefits analysis
benefit_keywords = {
    'effective': 0,
    'works well': 0,
    'great': 0,
    'excellent': 0,
    'best': 0,
    'lifesaver': 0,
    'no side effects': 0,
    'perfect': 0,
    'highly recommend': 0
}

for review in positive_reviews:
    if pd.notna(review):
        review_lower = str(review).lower()
        for benefit in benefit_keywords:
            if benefit in review_lower:
                benefit_keywords[benefit] += 1

sorted_benefits = sorted(benefit_keywords.items(), key=lambda x: x[1], reverse=True)

print(f"\n\nBenefits (from {len(positive_reviews)} positive reviews):")
print(f"{'Benefit Type':<25} {'Count':<10} {'Percentage'}")
print("-" * 60)

benefit_data = []
total_pos = len(positive_reviews)
for benefit, count in sorted_benefits[:10]:
    if count > 0:
        percentage = (count / total_pos) * 100
        print(f"{benefit:<25} {count:<10} {percentage:>6.2f}%")
        benefit_data.append({'Benefit': benefit, 'Count': count, 'Percentage': percentage})

benefit_df = pd.DataFrame(benefit_data)  # FIX: Was benefit_df, now benefit_data
```

* ### Can sentiment analysis predict drug ratings (1-10 scale) from review text alone? Why It Matters: Enables automated quality scoring without explicit ratings.

```python
# Analysis by drug name
drug_stats = drugs.groupby('drugName').agg({
    'rating': ['mean', 'count', 'std'],
    'sentiment_score': 'mean',
    'sentiment_label': lambda x: (x == 'POSITIVE').sum()
}).round(2)

drug_stats.columns = ['Avg_Rating', 'Review_Count', 'Std_Dev', 'Avg_Sentiment', 'Positive_Count']
drug_stats['Positive_Percent'] = (drug_stats['Positive_Count'] / drug_stats['Review_Count'] * 100).round(1)
drug_stats = drug_stats.sort_values('Avg_Rating', ascending=False)

print("Top 10 Drugs by Average Rating:")
print(drug_stats.head(10).to_string())

# Analysis by condition
condition_stats = drugs.groupby('condition').agg({
    'rating': ['mean', 'count'],
    'sentiment_score': 'mean',
    'sentiment_label': lambda x: (x == 'POSITIVE').sum()
}).round(2)

condition_stats.columns = ['Avg_Rating', 'Review_Count', 'Avg_Sentiment', 'Positive_Count']
condition_stats['Positive_Percent'] = (condition_stats['Positive_Count'] / condition_stats['Review_Count'] * 100).round(1)
condition_stats = condition_stats.sort_values('Avg_Rating', ascending=False)

print("\n\nTop 10 Conditions by Average Rating:")
print(condition_stats.head(10).to_string())

# Best and worst
best_drug = drug_stats.head(1)
worst_drug = drug_stats.tail(1)

print("\n" + "="*60)
print("KEY FINDINGS:")
print("="*60)
print(f"\n BEST PERFORMING DRUG:")
print(f"   {best_drug.index[0]}")
print(f"   Avg Rating: {best_drug['Avg_Rating'].values[0]}/10")
print(f"   {best_drug['Positive_Percent'].values[0]:.0f}% positive reviews")

print(f"\n  WORST PERFORMING DRUG:")
print(f"   {worst_drug.index[0]}")
print(f"   Avg Rating: {worst_drug['Avg_Rating'].values[0]}/10")
print(f"   {worst_drug['Positive_Percent'].values[0]:.0f}% positive reviews")
```

* ###  Can we detect adverse drug reactions (ADRs) early by monitoring negative sentiment spikes over time?

```python
import nltk
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Install if needed: pip install nrclex text2emotion

from nrclex import NRCLex

def detect_emotions(text):
    if pd.isna(text):
        return {'fear': 0, 'anger': 0, 'trust': 0, 'joy': 0, 'sadness': 0, 'anticipation': 0}
    
    emotion = NRCLex(str(text))
    return emotion.affect_frequencies

# Apply to reviews
drugs['emotions'] = drugs['review'].apply(detect_emotions)

# Extract individual emotions
drugs['fear'] = drugs['emotions'].apply(lambda x: x.get('fear', 0))
drugs['anger'] = drugs['emotions'].apply(lambda x: x.get('anger', 0))
drugs['trust'] = drugs['emotions'].apply(lambda x: x.get('trust', 0))  # Similar to "hope"
drugs['joy'] = drugs['emotions'].apply(lambda x: x.get('joy', 0))      # Similar to "relief"
drugs['sadness'] = drugs['emotions'].apply(lambda x: x.get('sadness', 0))
drugs['anticipation'] = drugs['emotions'].apply(lambda x: x.get('anticipation', 0))

# Get dominant emotion
def get_dominant_emotion(text):
    if pd.isna(text):
        return 'none'
    emotions = NRCLex(str(text)).top_emotions
    return emotions[0][0] if emotions else 'none'

drugs['dominant_emotion'] = drugs['review'].apply(get_dominant_emotion)

# Statistics
emotion_counts = drugs['dominant_emotion'].value_counts()
print("Emotion Distribution:")
print(emotion_counts)

# Compare emotions by sentiment
emotion_by_sentiment = drugs.groupby('sentiment_label')[['fear', 'anger', 'trust', 'joy']].mean()
print("\nAverage Emotion Scores by Sentiment:")
emotion_by_sentiment.head()
```

* ### How has patient sentiment about specific drug categories (e.g., antidepressants, diabetes medications) changed over time?

```python
# Calculate standard deviation of ratings (high = polarizing)
drug_polarization = drugs.groupby('drugName').agg({
    'rating': ['mean', 'std', 'count'],
    'sentiment_label': lambda x: x.value_counts().to_dict()
}).round(2)

drug_polarization.columns = ['Avg_Rating', 'Rating_StdDev', 'Review_Count', 'Sentiment_Distribution']

# Filter drugs with sufficient reviews
drug_polarization = drug_polarization[drug_polarization['Review_Count'] >= 50]

# Sort by standard deviation (most polarizing)
drug_polarization = drug_polarization.sort_values('Rating_StdDev', ascending=False)

print("Top 15 Most Polarizing Drugs (High Rating Variance):")
print(drug_polarization.head(15))

# Calculate polarization score
def calculate_polarization(group):
    positive = (group['sentiment_label'] == 'POSITIVE').sum()
    negative = (group['sentiment_label'] == 'NEGATIVE').sum()
    total = len(group)
    
    # High polarization when both positive and negative are significant
    polarization = min(positive, negative) / total * 2  # 0 to 1 scale
    return polarization

drug_polarization_score = drugs.groupby('drugName').apply(calculate_polarization)
drug_polarization_score = drug_polarization_score.sort_values(ascending=False)

print("\n\nTop 15 Drugs by Polarization Score (Mixed Positive/Negative):")
print(drug_polarization_score.head(15))

```

* ### Which drug categories (antidepressants, pain relief, diabetes) have the highest vs. lowest patient satisfaction?

```python
def extract_features(text):
    if pd.isna(text):
        return {
            'word_count': 0,
            'sentiment_score': 0,
            'positive_words': 0,
            'negative_words': 0,
            'has_side_effects': 0,
            'has_excellent': 0,
            'has_terrible': 0
        }
    
    text_lower = str(text).lower()
    
    features = {
        'word_count': len(str(text).split()),
        'sentiment_score': simple_sentiment_score(text),
        'positive_words': sum(1 for word in positive_words if word in text_lower),
        'negative_words': sum(1 for word in negative_words if word in text_lower),
        'has_side_effects': 1 if 'side effect' in text_lower else 0,
        'has_excellent': 1 if 'excellent' in text_lower else 0,
        'has_terrible': 1 if 'terrible' in text_lower else 0
    }
    return features

# Extract features
print("Extracting features from reviews...")
feature_list = []
for idx, row in drugs.head(1000).iterrows():
    features = extract_features(row['review'])
    features['actual_rating'] = row['rating']
    feature_list.append(features)

features_df = pd.DataFrame(feature_list)

# Simple prediction
features_df['predicted_simple'] = (features_df['sentiment_score'] + 1) * 5

# Advanced prediction
def predict_advanced(row):
    pred = 5.0
    pred += row['sentiment_score'] * 3.0
    pred += row['positive_words'] * 0.8
    pred -= row['negative_words'] * 0.8
    pred += row['has_excellent'] * 1.5
    pred -= row['has_terrible'] * 2.0
    pred -= row['has_side_effects'] * 0.5
    return max(1, min(10, pred))

features_df['predicted_advanced'] = features_df.apply(predict_advanced, axis=1)

# Calculate errors
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

mae_simple = mean_absolute_error(features_df['actual_rating'], features_df['predicted_simple'])
rmse_simple = math.sqrt(mean_squared_error(features_df['actual_rating'], features_df['predicted_simple']))

mae_advanced = mean_absolute_error(features_df['actual_rating'], features_df['predicted_advanced'])
rmse_advanced = math.sqrt(mean_squared_error(features_df['actual_rating'], features_df['predicted_advanced']))

print("\nModel Performance:")
print("="*60)
print(f"Simple Model (Sentiment Only):")
print(f"  MAE:  {mae_simple:.2f} points")
print(f"  RMSE: {rmse_simple:.2f} points")

print(f"\nAdvanced Model (Multi-Feature):")
print(f"  MAE:  {mae_advanced:.2f} points")
print(f"  RMSE: {rmse_advanced:.2f} points")
print(f"  Improvement: {((mae_simple - mae_advanced) / mae_simple * 100):.1f}%")

# Results
features_df.head()

```
