```python
import requests
from bs4 import BeautifulSoup

# Use a genuine-looking browser user agent
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/115.0.0.0 Safari/537.36"
    ),
    "Accept-Language": "en-US,en;q=0.9"
}

# Amazon search URL
url = "https://www.amazon.in/s?k=7000mh+battery+mobile+phones+5g"

# Send HTTP request
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

# Find all product containers
products = soup.select("div.s-main-slot div.s-result-item[data-component-type='s-search-result']")

# Parse and print details
for idx, product in enumerate(products, 1):
    try:
        name = product.select_one("span.a-text-normal").get_text().strip()
    except:
        name = "N/A"

    try:
        price = product.select_one("span.a-price-whole").get_text().strip()
    except:
        price = "N/A"

    try:
        rating = product.select_one("span.a-icon-alt").get_text().strip()
    except:
        rating = "No rating"

    try:
        img_url = product.select_one("img.s-image")["src"]
    except:
        img_url = "N/A"

    try:
        link = "https://www.amazon.in/s?k=laptops" + product.select_one("a.a-link-normal")["href"]
    except:
        link = "N/A"

    # Print the product info
    print(f"\nProduct {idx}")
    print(f"Name     : {name}")
    print(f"Price    : {price}")
    print(f"Rating   : {rating}")
    print(f"Image URL: {img_url}")
    print(f"Link     : {link}")
```
